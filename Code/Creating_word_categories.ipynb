## Check if words are present in all the models

```python
# Folder path where models are stored
models_folder = '/work/NLP_exam/he_she_aligned_models'

# Function to load all models in the folder
def load_models_from_folder(folder_path):
    models = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".w2v"): 
            model_path = os.path.join(folder_path, filename)
            model = Word2Vec.load(model_path)
            models[filename] = model.wv
    return models

# Function to check word presence in a model
def check_words_in_model(word_vectors, word_list):
    missing_words = [word for word in word_list if word not in word_vectors]
    return missing_words

# Load all models
models = load_models_from_folder(models_folder)

# Check word presence in all models
for model_name, word_vectors in models.items():
    print(f"Checking model: {model_name}")
    

    missing_high_agency = check_words_in_model(word_vectors, positive_agency)
    if missing_high_agency:
        print(f"Missing High Agency Words: {missing_high_agency}")
    

    missing_low_agency = check_words_in_model(word_vectors, negative_agency)
    if missing_low_agency:
        print(f"Missing Low Agency Words: {missing_low_agency}")
    

    missing_high_competence = check_words_in_model(word_vectors, high_competence)
    if missing_high_competence:
        print(f"Missing High Competence Words: {missing_high_competence}")
    

    missing_low_competence = check_words_in_model(word_vectors, low_competence)
    if missing_low_competence:
        print(f"Missing Low Competence Words: {missing_low_competence}")
    

    missing_high_warmth = check_words_in_model(word_vectors, high_warmth)
    if missing_high_warmth:
        print(f"Missing High Warmth Words: {missing_high_warmth}")
    

    missing_low_warmth = check_words_in_model(word_vectors, low_warmth)
    if missing_low_warmth:
        print(f"Missing Low Warmth Words: {missing_low_warmth}")


    missing_occ_female = check_words_in_model(word_vectors, occupation_trad_female)
    if missing_occ_female:
        print(f"Missing female occ Words: {missing_occ_female}")
    

    missing_occ_male = check_words_in_model(word_vectors, occupation_trad_male)
    if missing_occ_male:
        print(f"Missing male occ Words: {missing_occ_male}")


    missing_app_related = check_words_in_model(word_vectors, appearance_related)
    if missing_app_related:
        print(f"Missing appearance related Words: {missing_app_related}")
    
    print("-" * 40)
```
