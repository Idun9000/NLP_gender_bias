# Remove &-signs

```python
import os

def remove_ampersand(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        content = file.read()

    # Remove all occurrences of '&'
    content = content.replace('&', '')

    with open(filepath, 'w', encoding='utf-8') as file:
        file.write(content)
    print(f"Removed '&' from {filepath}")

# Define the folder containing TEI files
data_folder = "/Users/tildeidunsloth/Desktop/chicago_corpus/texts_renamed"

# Process each file
for filename in os.listdir(data_folder):
    if filename.endswith(".tei"):
        filepath = os.path.join(data_folder, filename)
        try:
            remove_ampersand(filepath)
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

# Rename files with year

```python
from lxml import etree
```

```python
def extract_year(filepath):
    try:
        # Parse the XML file
        tree = etree.parse(filepath)
        
        # Register the namespace
        namespaces = {'tei': 'http://www.tei-c.org/ns/1.0'}
        
        # Use the namespace-aware XPath
        year = tree.xpath("//tei:publicationStmt/tei:date/text()", namespaces=namespaces)
        if year:
            return year[0].strip()  # Return the first match
        else:
            print(f"No publication year found in {filepath}")
            return None
    except Exception as e:
        print(f"Error processing {filepath}: {e}")
        return None

def rename_file(filepath, year):
    if year:
        # Get the directory and original filename
        directory, original_filename = os.path.split(filepath)
        
        # Create the new filename by appending the year
        new_filename = original_filename.replace(".tei", f"Y_{year}.tei")
        
        # Create the full path for the new file
        new_filepath = os.path.join(directory, new_filename)
        
        # Rename the file
        os.rename(filepath, new_filepath)
        print(f"Renamed {original_filename} to {new_filename}")
    else:
        print(f"Year not found, so file {filepath} was not renamed.")
```

```python
# Loop over all files in the directory
for filename in os.listdir(data_folder):
    # Only process .tei files
    if filename.endswith(".tei"):
        filepath = os.path.join(data_folder, filename)
        
        # Extract the publication year
        year = extract_year(filepath)
        
        # Rename the file with the year if available
        rename_file(filepath, year)
```
